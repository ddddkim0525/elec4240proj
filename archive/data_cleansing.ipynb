```python
from PIL import Image
from numpy import asarray
import os

def load_dir(directory,target_directory):
    for texture_directory in os.listdir(directory):
        cnt = 0
        img_list = []
        for img_name in os.listdir(directory + "\\" + texture_directory):
            img_direc = directory + "\\" + texture_directory + "\\" + img_name
            save_direc = target_directory + "\\" 
            if img_name.lower().endswith("diff_1k.jpg"):
                save_direc += "diff"
            elif img_name.lower().endswith("nor_1k.jpg"):
                save_direc += "nor"
            elif img_name.lower().endswith("rough_1k.jpg"):
                save_direc += "rough"
            elif img_name.lower().endswith("disp_1k.jpg"):
                save_direc += "disp"
            if save_direc != (target_directory + "\\" ):
                Image.open(img_direc).save(save_direc+"\\"+img_name)
```
